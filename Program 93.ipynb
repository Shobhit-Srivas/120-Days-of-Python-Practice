# Problem 93: Create a script that takes a sentence and removes all stop words.

```python
print("=" * 60)
print("Q23. Create a script that takes a sentence and removes all stop words.")
print("=" * 60)

sentence = input("Enter a sentence: ")

stop_words = {"a", "an", "the", "is", "are", "was", "were", "in", "on", "at", "of", "to"}

result = [word for word in sentence.split() if word.lower() not in stop_words]

print("\nSentence After Removing Stop Words:")
print(" ".join(result))
```
