# Problem 119: Write a script to clean a text file by removing extra spaces and blank lines.

```python
print("=" * 60)
print("Q49. Write a script to clean a text file by removing extra spaces and blank lines.")
print("=" * 60)

file_name = input("Enter file name: ")

with open(file_name, "r") as file:
    lines = file.readlines()

cleaned = [line.strip() for line in lines if line.strip()]

with open(file_name, "w") as file:
    file.write("\n".join(cleaned))

print("\nFile Cleaned Successfully.")
```
