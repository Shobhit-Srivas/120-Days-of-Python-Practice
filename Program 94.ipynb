# Problem 94:  Write a Python program to split a sentence into words and reverse each word.

```python
print("=" * 60)
print("Q24. Write a Python program to split a sentence into words and reverse each word.")
print("=" * 60)

sentence = input("Enter a sentence: ")

result = [word[::-1] for word in sentence.split()]

print("\nReversed Words:")
print(" ".join(result))
```
