# Problem 73: Write a script to count the frequency of each character in a string.

```python
print("=" * 60)
print("Q3. Write a script to count the frequency of each character in a string.")
print("=" * 60)

text = input("Enter a string: ")

freq = {}

for ch in text:
    freq[ch] = freq.get(ch, 0) + 1

print("\nCharacter Frequencies:")
print(freq)
```
