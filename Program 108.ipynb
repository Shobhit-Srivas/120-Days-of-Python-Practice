# Problem 108: Create a function that removes all HTML tags from a string.

```python
print("=" * 60)
print("Q38. Create a function that removes all HTML tags from a string.")
print("=" * 60)

import re

text = input("Enter HTML content: ")

result = re.sub(r"<.*?>", "", text)

print("\nText Without HTML Tags:")
print(result)
```
