# Problem 56: Create a functon to rotate a list lef by k positons.

```python
print("=" * 60)
print("Q26. Create a function to rotate a list left by k positions.")
print("=" * 60)

def rotate_left(lst, k):
    return lst[k:] + lst[:k]

numbers = list(map(int, input("Enter list elements: ").split()))
k = int(input("Enter number of positions to rotate: "))

print("\nRotated List:")
print(rotate_left(numbers, k))
```
