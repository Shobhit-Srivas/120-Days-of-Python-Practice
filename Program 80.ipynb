# Problem 80:  Write a Python program to capitalize the first letter of each word in a sentence.

```python
print("=" * 60)
print("Q10. Write a Python program to capitalize the first letter of each word in a sentence.")
print("=" * 60)

sentence = input("Enter a sentence: ")

result = " ".join(word.capitalize() for word in sentence.split())

print("\nCapitalized Sentence:")
print(result)
```
