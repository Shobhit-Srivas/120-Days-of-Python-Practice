# Problem 84: Write a script to convert a string into title case without using .title().

```python
print("=" * 60)
print("Q14. Write a script to convert a string into title case without using .title().")
print("=" * 60)

text = input("Enter a sentence: ")

result = " ".join(word[:1].upper() + word[1:].lower() for word in text.split())

print("\nTitle Case String:")
print(result)
```
