# Problem 104: Write a script to check if a file contains a specific word.

```python
print("=" * 60)
print("Q34. Write a script to check if a file contains a specific word.")
print("=" * 60)

file_name = input("Enter file name: ")
word = input("Enter word to search: ")

with open(file_name, "r") as file:
    content = file.read()

if word in content:
    print("\nWord Found in File.")
else:
    print("\nWord Not Found in File.")
```
