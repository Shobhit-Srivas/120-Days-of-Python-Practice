# Problem 30: Write a program to display the cube of the number up to an integer

```python
print("=" * 60)
print("Q30. Write a program to display the cube of the number up to an integer.")
print("=" * 60)

n = int(input("Enter an integer: "))

print("\nNumber\tCube")

for i in range(1, n + 1):
    print(i, "\t", i ** 3)
```
