# Problem 98: Write a Python program to count the frequency of each word in a file.

```python
print("=" * 60)
print("Q28. Write a Python program to count the frequency of each word in a file.")
print("=" * 60)

file_name = input("Enter file name: ")

with open(file_name, "r") as file:
    words = file.read().split()

frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

print("\nWord Frequencies:")
print(frequency)
```
