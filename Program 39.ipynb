# Problem 39: Write a functon to count the frequency of elements in a list.

```python
print("=" * 60)
print("Q9. Write a function to count the frequency of elements in a list.")
print("=" * 60)

def frequency(lst):
    result = {}

    for item in lst:
        result[item] = result.get(item, 0) + 1

    return result

numbers = input("Enter elements separated by spaces: ").split()

print("\nElement Frequencies:")
print(frequency(numbers))
```
