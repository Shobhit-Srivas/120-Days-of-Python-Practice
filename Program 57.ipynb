# Problem 57: Write a functon to fnd the missing number from a list of 1 to N.

```python
print("=" * 60)
print("Q27. Write a function to find the missing number from a list of 1 to N.")
print("=" * 60)

def find_missing(numbers, n):
    return n * (n + 1) // 2 - sum(numbers)

numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))
n = int(input("Enter the value of N: "))

print("\nMissing Number:", find_missing(numbers, n))
```
