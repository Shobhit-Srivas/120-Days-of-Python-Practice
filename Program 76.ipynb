# Problem 76:  Write a program that takes a string and returns the string in reverse order without using [::-1]

```python
print("=" * 60)
print("Q6. Write a program that takes a string and returns the string in reverse order without using [::-1].")
print("=" * 60)

text = input("Enter a string: ")

reversed_text = ""

for ch in text:
    reversed_text = ch + reversed_text

print("\nReversed String:")
print(reversed_text)
```
