# Problem 14: Check if a given year is a leap year or not.

```python
print("=" * 60)
print("Q14. Check if a given year is a leap year or not.")
print("=" * 60)

year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("\n", year, "is a Leap Year")
else:
    print("\n", year, "is Not a Leap Year")
```
