# Problem 110: Write a script that finds all email addresses in a given text.

```python
print("=" * 60)
print("Q40. Write a script that finds all email addresses in a given text.")
print("=" * 60)

import re

text = input("Enter text: ")

emails = re.findall(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", text)

print("\nEmail Addresses Found:")
print(emails)
```
