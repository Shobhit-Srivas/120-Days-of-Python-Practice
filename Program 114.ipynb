# Problem 114: Write a program that extracts all integers from a given text.

```python
print("=" * 60)
print("Q44. Write a program that extracts all integers from a given text.")
print("=" * 60)

import re

text = input("Enter text: ")

numbers = re.findall(r"-?\d+", text)

print("\nExtracted Integers:")
print(numbers)
```
