# Problem 48: Create a dictonary and iterate over its keys and values.

```python
print("=" * 60)
print("Q18. Create a dictionary and iterate over its keys and values.")
print("=" * 60)

data = {
    "Name": "Shobhit",
    "Age": 21,
    "City": "Prayagraj"
}

print("\nDictionary Contents:")

for key, value in data.items():
    print(key, ":", value)
```
