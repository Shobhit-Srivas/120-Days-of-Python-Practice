# Problem 88: Create a program that counts words, characters, and lines in a paragraph.

```python
print("=" * 60)
print("Q18. Create a program that counts words, characters, and lines in a paragraph.")
print("=" * 60)

paragraph = input("Enter a paragraph: ")

words = len(paragraph.split())
characters = len(paragraph)
lines = len(paragraph.splitlines())

print("\nNumber of Words:", words)
print("Number of Characters:", characters)
print("Number of Lines:", lines)
```
