# Problem 23:  Write a Python program to print a pattern of stars in a triangle.

```python
print("=" * 60)
print("Q23. Write a Python program to print a pattern of stars in a triangle.")
print("=" * 60)

rows = int(input("Enter the number of rows: "))

print("\nStar Pattern:\n")

for i in range(1, rows + 1):
    print("*" * i)
```
