# Problem 117: Write a script that finds the longest sentence in a paragraph.

```python
print("=" * 60)
print("Q47. Write a script that finds the longest sentence in a paragraph.")
print("=" * 60)

import re

paragraph = input("Enter a paragraph: ")

sentences = re.split(r"[.!?]", paragraph)

longest = max(sentences, key=len)

print("\nLongest Sentence:")
print(longest.strip())
```
