# Problem 54: Write a functon to faten a nested list.

```python
print("=" * 60)
print("Q24. Write a function to flatten a nested list.")
print("=" * 60)

def flatten(lst):
    result = []

    for item in lst:
        if isinstance(item, list):
            result.extend(item)
        else:
            result.append(item)

    return result

nested_list = [[1, 2], [3, 4], [5, 6]]

print("\nOriginal List:")
print(nested_list)

print("\nFlattened List:")
print(flatten(nested_list))
```
