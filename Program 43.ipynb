# Problem 43: Create a set and perform union, intersecton, and diference.

```python
print("=" * 60)
print("Q13. Create a set and perform union, intersection, and difference.")
print("=" * 60)

set1 = set(map(int, input("Enter elements of Set 1: ").split()))
set2 = set(map(int, input("Enter elements of Set 2: ").split()))

print("\nUnion:")
print(set1 | set2)

print("\nIntersection:")
print(set1 & set2)

print("\nDifference (Set1 - Set2):")
print(set1 - set2)
```
