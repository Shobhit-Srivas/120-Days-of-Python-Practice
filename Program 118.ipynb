# Problem 118: Write a Python program to read a file and display all lines that contain a given keyword.

```python
print("=" * 60)
print("Q48. Write a Python program to read a file and display all lines that contain a given keyword.")
print("=" * 60)

file_name = input("Enter file name: ")
keyword = input("Enter keyword: ")

with open(file_name, "r") as file:
    print("\nMatching Lines:\n")

    for line in file:
        if keyword.lower() in line.lower():
            print(line.strip())
```
