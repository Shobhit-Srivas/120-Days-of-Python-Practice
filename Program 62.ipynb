```python
print("=" * 60)
print("Q2. Create a file and write your name into it.")
print("=" * 60)

file_name = input("Enter file name: ")
name = input("Enter your name: ")

with open(file_name, "w") as file:
    file.write(name)

print("\nName successfully written to the file.")
```
