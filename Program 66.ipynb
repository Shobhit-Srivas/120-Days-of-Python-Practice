# Problem 66: Use a list comprehension to flter even numbers from a list.

```python
print("=" * 60)
print("Q6. Use a list comprehension to filter even numbers from a list.")
print("=" * 60)

numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))

even_numbers = [num for num in numbers if num % 2 == 0]

print("\nEven Numbers:")
print(even_numbers)
```
