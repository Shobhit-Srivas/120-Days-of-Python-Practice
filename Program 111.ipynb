# Problem 111: Write a program that counts the occurrence of each vowel in a paragraph.

```python
print("=" * 60)
print("Q41. Write a program that counts the occurrence of each vowel in a paragraph.")
print("=" * 60)

text = input("Enter a paragraph: ").lower()

vowels = {"a": 0, "e": 0, "i": 0, "o": 0, "u": 0}

for ch in text:
    if ch in vowels:
        vowels[ch] += 1

print("\nVowel Occurrences:")
print(vowels)
```
