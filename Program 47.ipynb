# Problem 47: Write a functon to count vowels in a string.

```python
print("=" * 60)
print("Q17. Write a function to count vowels in a string.")
print("=" * 60)

def count_vowels(text):
    return sum(1 for ch in text.lower() if ch in "aeiou")

text = input("Enter a string: ")

print("\nNumber of vowels:", count_vowels(text))
```
