# Problem 81: Create a program that extracts numbers from a string and returns their sum.

```python
print("=" * 60)
print("Q11. Create a program that extracts numbers from a string and returns their sum.")
print("=" * 60)

import re

text = input("Enter a string: ")

numbers = re.findall(r'\d+', text)

total = sum(map(int, numbers))

print("\nSum of Numbers:", total)
```
