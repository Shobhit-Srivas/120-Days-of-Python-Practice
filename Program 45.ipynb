# Problem 45: Write a functon that returns the factorial of a number.

```python
print("=" * 60)
print("Q15. Write a function that returns the factorial of a number.")
print("=" * 60)

def factorial(n):
    result = 1

    for i in range(1, n + 1):
        result *= i

    return result

num = int(input("Enter a number: "))

print("\nFactorial:", factorial(num))
```
