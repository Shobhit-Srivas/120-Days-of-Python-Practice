# Problem 97:  Write a function that compresses a string using run-length encoding.

```python
print("=" * 60)
print("Q27. Write a function that compresses a string using run-length encoding.")
print("=" * 60)

text = input("Enter a string: ")

result = ""
count = 1

for i in range(1, len(text)):
    if text[i] == text[i - 1]:
        count += 1
    else:
        result += text[i - 1] + str(count)
        count = 1

result += text[-1] + str(count)

print("\nCompressed String:")
print(result)
```
