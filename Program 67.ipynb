# Problem 67: Write a generator that yields even numbers up to N.

```python
print("=" * 60)
print("Q7. Write a generator that yields even numbers up to N.")
print("=" * 60)

def even_generator(n):
    for i in range(2, n + 1, 2):
        yield i

n = int(input("Enter the value of N: "))

print("\nEven Numbers:")

for num in even_generator(n):
    print(num, end=" ")
```
