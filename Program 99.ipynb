# Problem 99: Write a script that extracts hashtags from a tweet.

```python
print("=" * 60)
print("Q29. Write a script that extracts hashtags from a tweet.")
print("=" * 60)

import re

tweet = input("Enter a tweet: ")

hashtags = re.findall(r"#\w+", tweet)

print("\nExtracted Hashtags:")
print(hashtags)
```
