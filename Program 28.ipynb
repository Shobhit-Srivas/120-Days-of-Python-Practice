# Problem 28: Write a program to find the sum of all even numbers in a list.

```python
print("=" * 60)
print("Q28. Write a program to find the sum of all even numbers in a list.")
print("=" * 60)

numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))

even_sum = sum(num for num in numbers if num % 2 == 0)

print("\nSum of all even numbers:", even_sum)
```
