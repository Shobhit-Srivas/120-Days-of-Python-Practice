# Problem 44: Write a functon to fnd common elements in two lists.

```python
print("=" * 60)
print("Q14. Write a function to find common elements in two lists.")
print("=" * 60)

def common_elements(a, b):
    return list(set(a) & set(b))

list1 = list(map(int, input("Enter elements of List 1: ").split()))
list2 = list(map(int, input("Enter elements of List 2: ").split()))

print("\nCommon Elements:")
print(common_elements(list1, list2))
```
