# Problem 15: Create a program to print the Fibonacci series up to N terms.

```python
print("=" * 60)
print("Q15. Create a program to print the Fibonacci series up to N terms.")
print("=" * 60)

n = int(input("Enter the number of terms: "))

a = 0
b = 1

print("\nFibonacci Series:")

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```
