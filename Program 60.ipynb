# Problem 60: Create a functon to fnd unique elements present in only one of two lists.

```python
print("=" * 60)
print("Q30. Create a function to find unique elements present in only one of two lists.")
print("=" * 60)

def unique_elements(list1, list2):
    return list(set(list1) ^ set(list2))

list1 = list(map(int, input("Enter elements of List 1: ").split()))
list2 = list(map(int, input("Enter elements of List 2: ").split()))

print("\nUnique Elements:")
print(unique_elements(list1, list2))
```
