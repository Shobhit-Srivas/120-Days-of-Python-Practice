# Problem 61: Write a Python script to read a fle and print its contents.

```python
print("=" * 60)
print("Q1. Write a Python script to read a file and print its contents.")
print("=" * 60)

file_name = input("Enter file name: ")

with open(file_name, "r") as file:
    content = file.read()

print("\nFile Contents:\n")
print(content)
```
