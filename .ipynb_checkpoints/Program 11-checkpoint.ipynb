# Problem 11:  Write a program to find the sum of first N natural numbers.

```python
print("=" * 60)
print("Q11. Write a program to find the sum of first N natural numbers.")
print("=" * 60)

n = int(input("Enter a positive integer: "))

total = n * (n + 1) // 2

print("\nSum of first", n, "natural numbers is:", total)
```
