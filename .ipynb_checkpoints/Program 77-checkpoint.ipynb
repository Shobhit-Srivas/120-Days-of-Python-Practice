# Problem 77:  Create a Python function to check if a string is a pangram.

```python
print("=" * 60)
print("Q7. Create a Python function to check if a string is a pangram.")
print("=" * 60)

def is_pangram(text):
    alphabet = set("abcdefghijklmnopqrstuvwxyz")
    return alphabet <= set(text.lower())

text = input("Enter a sentence: ")

if is_pangram(text):
    print("\nThe sentence is a Pangram.")
else:
    print("\nThe sentence is not a Pangram.")
```
