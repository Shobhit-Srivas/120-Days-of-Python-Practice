# Problem 59: Write a functon to merge two dictonaries and handle key collisions by summing values.

```python
print("=" * 60)
print("Q29. Write a function to merge two dictionaries and handle key collisions by summing values.")
print("=" * 60)

def merge_dicts(d1, d2):
    result = d1.copy()

    for key, value in d2.items():
        result[key] = result.get(key, 0) + value

    return result

dict1 = {"a": 10, "b": 20}
dict2 = {"b": 30, "c": 40}

print("\nMerged Dictionary:")
print(merge_dicts(dict1, dict2))
```
