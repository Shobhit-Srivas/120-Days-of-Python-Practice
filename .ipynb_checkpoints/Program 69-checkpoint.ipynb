# Problem 69: Write a program to read a CSV fle and print its contents.

```python
import csv

print("=" * 60)
print("Q9. Write a program to read a CSV file and print its contents.")
print("=" * 60)

file_name = input("Enter CSV file name: ")

with open(file_name, "r") as file:
    reader = csv.reader(file)

    print("\nCSV Contents:\n")

    for row in reader:
        print(row)
```
