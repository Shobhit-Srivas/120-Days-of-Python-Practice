# Program 105: Write a Python program to find and replace text in a file.

```python
print("=" * 60)
print("Q35. Write a Python program to find and replace text in a file.")
print("=" * 60)

file_name = input("Enter file name: ")
old_text = input("Enter text to find: ")
new_text = input("Enter replacement text: ")

with open(file_name, "r") as file:
    content = file.read()

content = content.replace(old_text, new_text)

with open(file_name, "w") as file:
    file.write(content)

print("\nText Replaced Successfully.")
```
