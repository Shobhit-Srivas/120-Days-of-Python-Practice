# Problem 96: Write a program to find all palindromic substrings in a string.

```python
print("=" * 60)
print("Q26. Write a program to find all palindromic substrings in a string.")
print("=" * 60)

text = input("Enter a string: ")

palindromes = set()

for i in range(len(text)):
    for j in range(i + 1, len(text) + 1):
        sub = text[i:j]
        if len(sub) > 1 and sub == sub[::-1]:
            palindromes.add(sub)

print("\nPalindromic Substrings:")
print(list(palindromes))
```
