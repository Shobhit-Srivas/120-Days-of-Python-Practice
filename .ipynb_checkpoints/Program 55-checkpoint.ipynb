# Problem 55: Write a program to fnd the second highest number in a list.

```python
print("=" * 60)
print("Q25. Write a program to find the second highest number in a list.")
print("=" * 60)

numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))

unique_numbers = sorted(set(numbers), reverse=True)

print("\nSecond Highest Number:", unique_numbers[1])
```
