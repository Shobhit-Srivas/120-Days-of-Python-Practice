# Problem 92:  Write a program to count the number of words starting with a vowel in a string

```python
print("=" * 60)
print("Q22. Write a program to count the number of words starting with a vowel in a string.")
print("=" * 60)

sentence = input("Enter a sentence: ")

count = sum(1 for word in sentence.split() if word[0].lower() in "aeiou")

print("\nNumber of Words Starting with a Vowel:", count)
```
