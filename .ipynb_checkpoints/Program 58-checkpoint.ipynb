# Problem 58: Write a program to remove all None values from a list.

```python
print("=" * 60)
print("Q28. Write a program to remove all None values from a list.")
print("=" * 60)

data = [1, None, 2, None, 3, 4, None, 5]

result = [item for item in data if item is not None]

print("\nOriginal List:")
print(data)

print("\nList After Removing None Values:")
print(result)
```
