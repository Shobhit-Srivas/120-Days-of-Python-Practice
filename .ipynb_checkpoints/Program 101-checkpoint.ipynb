# Problem 101:  Create a program that finds the first non-repeating character in a string.

```python
print("=" * 60)
print("Q31. Create a program that finds the first non-repeating character in a string.")
print("=" * 60)

text = input("Enter a string: ")

for ch in text:
    if text.count(ch) == 1:
        print("\nFirst Non-Repeating Character:", ch)
        break
else:
    print("\nNo Non-Repeating Character Found.")
```
