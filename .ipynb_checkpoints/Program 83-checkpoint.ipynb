# Problem 83: Write a function to count how many times a substring appears in a string.

```python
print("=" * 60)
print("Q13. Write a function to count how many times a substring appears in a string.")
print("=" * 60)

text = input("Enter a string: ")
substring = input("Enter a substring: ")

count = text.count(substring)

print("\nNumber of Occurrences:", count)
```
