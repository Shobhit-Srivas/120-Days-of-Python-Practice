# Problem 31:  Write a function to check if a number is even.

```python
print("=" * 60)
print("Q1. Write a function to check if a number is even.")
print("=" * 60)

def is_even(n):
    return n % 2 == 0

num = int(input("Enter a number: "))

if is_even(num):
    print("\nResult:", num, "is an Even Number")
else:
    print("\nResult:", num, "is an Odd Number")
```
