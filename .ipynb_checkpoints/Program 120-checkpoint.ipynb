# Problem 120: Write a Python program to count how many sentences are in a paragraph.

```python
print("=" * 60)
print("Q50. Write a Python program to count how many sentences are in a paragraph.")
print("=" * 60)

import re

paragraph = input("Enter a paragraph: ")

sentences = re.findall(r"[^.!?]+[.!?]", paragraph)

print("\nNumber of Sentences:", len(sentences))
```
