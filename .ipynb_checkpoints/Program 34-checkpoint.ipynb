# Problem 34:  Create a program that removes duplicates from a list.

```python
print("=" * 60)
print("Q4. Create a program that removes duplicates from a list.")
print("=" * 60)

numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))

result = list(dict.fromkeys(numbers))

print("\nList after removing duplicates:")
print(result)
```
