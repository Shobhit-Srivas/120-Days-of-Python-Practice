# Problem 18: Check whether a character is a vowel or consonant.

```python
print("=" * 60)
print("Q18. Check whether a character is a vowel or consonant.")
print("=" * 60)

ch = input("Enter an alphabet: ").lower()

if ch in "aeiou":
    print("\n", ch, "is a Vowel")
else:
    print("\n", ch, "is a Consonant")
```
