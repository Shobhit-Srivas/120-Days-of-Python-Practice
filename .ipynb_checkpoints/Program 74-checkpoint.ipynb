# Problem 74: Write a program that accepts a sentence and calculates the number of upper and lower case leters.

```python
print("=" * 60)
print("Q4. Write a program that accepts a sentence and calculates the number of upper and lower case letters.")
print("=" * 60)

text = input("Enter a sentence: ")

upper = sum(1 for ch in text if ch.isupper())
lower = sum(1 for ch in text if ch.islower())

print("\nNumber of Uppercase Letters:", upper)
print("Number of Lowercase Letters:", lower)
```
