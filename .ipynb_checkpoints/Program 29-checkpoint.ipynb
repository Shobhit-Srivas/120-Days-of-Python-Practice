# Problem 29: Create a program to check whether a number is prime or not.

```python
print("=" * 60)
print("Q29. Write a program to check whether a number is prime or not.")
print("=" * 60)

num = int(input("Enter a number: "))

if num > 1:
    is_prime = True

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print("\n", num, "is a Prime Number")
    else:
        print("\n", num, "is Not a Prime Number")
else:
    print("\n", num, "is Not a Prime Number")
```
