# Problem 16: Write a program to find the GCD of two numbers.

```python
import math

print("=" * 60)
print("Q16. Write a program to find the GCD of two numbers.")
print("=" * 60)

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

gcd = math.gcd(a, b)

print("\nGCD of", a, "and", b, "is:", gcd)
```
