# Problem 89: Write a script to encode a string using Caesar cipher (shift = 3).

```python
print("=" * 60)
print("Q19. Write a script to encode a string using Caesar cipher (shift = 3).")
print("=" * 60)

text = input("Enter a string: ")

result = ""

for ch in text:
    if ch.isalpha():
        start = ord('A') if ch.isupper() else ord('a')
        result += chr((ord(ch) - start + 3) % 26 + start)
    else:
        result += ch

print("\nEncoded String:")
print(result)
```
