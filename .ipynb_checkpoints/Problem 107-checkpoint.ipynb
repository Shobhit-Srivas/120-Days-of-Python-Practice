# Problem 107: Write a program to calculate the average word length in a sentence.

```python
print("=" * 60)
print("Q37. Write a program to calculate the average word length in a sentence.")
print("=" * 60)

sentence = input("Enter a sentence: ")

words = sentence.split()

average = sum(len(word) for word in words) / len(words)

print("\nAverage Word Length:", round(average, 2))
```
