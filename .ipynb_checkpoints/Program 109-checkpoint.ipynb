# Problem 109:  Write a program to parse a date string and display it in a different format.

```python
print("=" * 60)
print("Q39. Write a program to parse a date string and display it in a different format.")
print("=" * 60)

from datetime import datetime

date_string = input("Enter date (DD-MM-YYYY): ")

date_object = datetime.strptime(date_string, "%d-%m-%Y")

print("\nFormatted Date:")
print(date_object.strftime("%Y/%m/%d"))
```
