# Problem 102:  Write a script that converts camelCase to snake_case.

```python
print("=" * 60)
print("Q32. Write a script that converts camelCase to snake_case.")
print("=" * 60)

import re

text = input("Enter a camelCase string: ")

result = re.sub(r'([A-Z])', r'_\1', text).lower()

print("\nSnake Case String:")
print(result)
```
