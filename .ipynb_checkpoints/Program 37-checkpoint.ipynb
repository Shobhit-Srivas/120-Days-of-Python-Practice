# Problem 37: Convert a list into a tuple and vice versa.

```python
print("=" * 60)
print("Q7. Convert a list into a tuple and vice versa.")
print("=" * 60)

lst = input("Enter list elements separated by spaces: ").split()

tpl = tuple(lst)

print("\nTuple:", tpl)

new_list = list(tpl)

print("List:", new_list)
```
