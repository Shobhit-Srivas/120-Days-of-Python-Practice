# Problem 100: Write a function to remove punctuation from a string.

```python
print("=" * 60)
print("Q30. Write a function to remove punctuation from a string.")
print("=" * 60)

import string

text = input("Enter a string: ")

result = "".join(ch for ch in text if ch not in string.punctuation)

print("\nString After Removing Punctuation:")
print(result)
```
