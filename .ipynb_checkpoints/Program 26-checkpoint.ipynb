# Problem 26:  Convert a decimal number to binary using loops.

```python
print("=" * 60)
print("Q26. Convert a decimal number to binary using loops.")
print("=" * 60)

num = int(input("Enter a decimal number: "))

binary = ""

if num == 0:
    binary = "0"
else:
    while num > 0:
        binary = str(num % 2) + binary
        num //= 2

print("\nBinary representation:", binary)
```
