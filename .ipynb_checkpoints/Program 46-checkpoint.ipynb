# Problem 46: Create a functon that checks whether a string is a palindrome.

```python
print("=" * 60)
print("Q16. Create a function that checks whether a string is a palindrome.")
print("=" * 60)

def is_palindrome(text):
    return text == text[::-1]

text = input("Enter a string: ")

if is_palindrome(text):
    print("\nThe string is a Palindrome.")
else:
    print("\nThe string is not a Palindrome.")
```
