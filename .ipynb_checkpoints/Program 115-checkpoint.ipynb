# Problem 115: Create a script to find duplicate words in a paragraph.

```python
print("=" * 60)
print("Q45. Create a script to find duplicate words in a paragraph.")
print("=" * 60)

text = input("Enter a paragraph: ").lower()

words = text.split()

duplicates = {word for word in words if words.count(word) > 1}

print("\nDuplicate Words:")
print(list(duplicates))
```
