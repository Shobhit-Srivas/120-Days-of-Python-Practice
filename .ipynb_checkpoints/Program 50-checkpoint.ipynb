# Problem 50: Write a functon to capitalize the frst leter of each word in a string.

```python
print("=" * 60)
print("Q20. Write a function to capitalize the first letter of each word in a string.")
print("=" * 60)

def capitalize_words(text):
    return " ".join(word.capitalize() for word in text.split())

text = input("Enter a sentence: ")

print("\nCapitalized Sentence:")
print(capitalize_words(text))
```
