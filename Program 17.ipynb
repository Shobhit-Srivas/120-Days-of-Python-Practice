# Porblem 17: Write a program to find the LCM of two numbers.

```python
import math

print("=" * 60)
print("Q17. Write a program to find the LCM of two numbers.")
print("=" * 60)

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

lcm = abs(a * b) // math.gcd(a, b)

print("\nLCM of", a, "and", b, "is:", lcm)
```
