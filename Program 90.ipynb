# Problem 90: Write a program that accepts a string and counts vowels and consonants.

```python
print("=" * 60)
print("Q20. Write a program that accepts a string and counts vowels and consonants.")
print("=" * 60)

text = input("Enter a string: ").lower()

vowels = 0
consonants = 0

for ch in text:
    if ch.isalpha():
        if ch in "aeiou":
            vowels += 1
        else:
            consonants += 1

print("\nNumber of Vowels:", vowels)
print("Number of Consonants:", consonants)
```
