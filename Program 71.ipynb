# Problem 71: Write a Python program to check if a string has all unique characters.

```python
print("=" * 60)
print("Q1. Write a Python program to check if a string has all unique characters.")
print("=" * 60)

text = input("Enter a string: ")

if len(text) == len(set(text)):
    print("\nAll characters are unique.")
else:
    print("\nThe string contains duplicate characters.")
```
