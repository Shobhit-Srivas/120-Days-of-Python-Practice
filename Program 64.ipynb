# Problem 64: Write a program to handle fle not found error.

```python
print("=" * 60)
print("Q4. Write a program to handle file not found error.")
print("=" * 60)

file_name = input("Enter file name: ")

try:
    with open(file_name, "r") as file:
        print("\nFile Contents:\n")
        print(file.read())

except FileNotFoundError:
    print("\nError: File not found.")
```
