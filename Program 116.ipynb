# Problem 116: Write a program that converts a sentence to Pig Latin.

```python
print("=" * 60)
print("Q46. Write a program that converts a sentence to Pig Latin.")
print("=" * 60)

sentence = input("Enter a sentence: ")

result = []

for word in sentence.split():
    result.append(word[1:] + word[0] + "ay")

print("\nPig Latin Sentence:")
print(" ".join(result))
```
