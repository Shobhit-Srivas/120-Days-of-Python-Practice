# Problem 103: Write a function to generate acronyms from a sentence.

```python
print("=" * 60)
print("Q33. Write a function to generate acronyms from a sentence.")
print("=" * 60)

sentence = input("Enter a sentence: ")

acronym = "".join(word[0].upper() for word in sentence.split())

print("\nAcronym:")
print(acronym)
```
