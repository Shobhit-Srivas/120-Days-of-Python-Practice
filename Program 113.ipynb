# Problem 113: Write a script to check if a string is a valid URL.

```python
print("=" * 60)
print("Q43. Write a script to check if a string is a valid URL.")
print("=" * 60)

import re

url = input("Enter a URL: ")

pattern = r"^(https?:\/\/)?([\w\-]+\.)+[\w\-]+(\/[\w\-./?%&=]*)?$"

if re.match(pattern, url):
    print("\nValid URL")
else:
    print("\nInvalid URL")
```
