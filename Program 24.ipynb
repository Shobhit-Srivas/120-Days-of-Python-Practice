# Problem 24:  Create a calculator app using if-else.

```python
print("=" * 60)
print("Q24. Create a calculator app using if-else.")
print("=" * 60)

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")

if operator == "+":
    result = num1 + num2
elif operator == "-":
    result = num1 - num2
elif operator == "*":
    result = num1 * num2
elif operator == "/":
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Division by zero is not allowed"
else:
    result = "Invalid operator"

print("\nResult:", result)
```
