# Problem 86: Create a program to filter out all non-alphabetic characters from a string.

```python
print("=" * 60)
print("Q16. Create a program to filter out all non-alphabetic characters from a string.")
print("=" * 60)

text = input("Enter a string: ")

result = "".join(ch for ch in text if ch.isalpha())

print("\nAlphabetic Characters Only:")
print(result)
```
