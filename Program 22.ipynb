# Problem 22: Create a program to print all Armstrong numbers between 1 to 1000.

```python
print("=" * 60)
print("Q22. Create a program to print all Armstrong numbers between 1 to 1000.")
print("=" * 60)

print("\nArmstrong Numbers between 1 and 1000:")

for num in range(1, 1001):
    total = sum(int(digit) ** len(str(num)) for digit in str(num))

    if total == num:
        print(num, end=" ")
```
