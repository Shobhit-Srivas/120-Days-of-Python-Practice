# Problem 112: Create a function that validates an email address format.

```python
print("=" * 60)
print("Q42. Create a function that validates an email address format.")
print("=" * 60)

import re

email = input("Enter an email address: ")

pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"

if re.match(pattern, email):
    print("\nValid Email Address")
else:
    print("\nInvalid Email Address")
```
