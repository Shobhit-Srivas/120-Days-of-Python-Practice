# Problem 51:  Create a list comprehension to get squares of all even numbers in a range.

```python
print("=" * 60)
print("Q21. Create a list comprehension to get squares of all even numbers in a range.")
print("=" * 60)

n = int(input("Enter the range limit: "))

squares = [i ** 2 for i in range(n + 1) if i % 2 == 0]

print("\nSquares of Even Numbers:")
print(squares)
```
