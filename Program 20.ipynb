# Problem 20: Create a program to find the second largest number in a list.

```python
print("=" * 60)
print("Q20. Create a program to find the second largest number in a list.")
print("=" * 60)

numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))

unique_numbers = list(set(numbers))
unique_numbers.sort(reverse=True)

print("\nSecond largest number is:", unique_numbers[1])
```
