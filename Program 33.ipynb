# Problem 33: Write a program to find the maximum and minimum in a list.

```python
print("=" * 60)
print("Q3. Write a program to find the maximum and minimum in a list.")
print("=" * 60)

numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))

print("\nMaximum value:", max(numbers))
print("Minimum value:", min(numbers))
```
