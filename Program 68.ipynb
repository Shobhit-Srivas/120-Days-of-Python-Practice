# Problem 68: Create a program to count lines and words in a fle.

```python
print("=" * 60)
print("Q8. Create a program to count lines and words in a file.")
print("=" * 60)

file_name = input("Enter file name: ")

with open(file_name, "r") as file:
    content = file.readlines()

line_count = len(content)
word_count = sum(len(line.split()) for line in content)

print("\nNumber of Lines:", line_count)
print("Number of Words:", word_count)
```
